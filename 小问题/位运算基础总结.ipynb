## 位运算基础总结

#### [python中的位运算总结](https://www.cnblogs.com/Neeo/articles/10536202.html)


#### & : 两个值都为1才为1

```python
print(3&6)
bin(3),bin(6),bin(3&6)
```

```python
True & False, True & True, True & True, False & False
```

#### |: 有一个为1就为1

```python
print(3|6)
bin(3),bin(6),bin(3|6)
```

```python
True | False, True | True, True | True, False | False
```

#### ^: 异或，两个不一样才为1

```python
print(3^6)
bin(3),bin(6),bin(3^6)
```

```python
True ^ False, True ^ True, True ^ True, False ^ False
```

#### ~: 取反

```python
print(~6)
bin(6),bin(~6)
```

```python
# 结果解释
# 6的二进制表示               0000 0110
# 取反操作                    1111 1001
# 高位不变，低位取反，加一    1000 0111
# 最后结果                    -7
```

#### <</>> 左移和右移（箭头朝那边就往那边移动）

```python
# 左移一位，变大两倍
6<<1,bin(6),bin(6<<1)
```

```python
# 右移一位，缩小两倍，向下取整
3>>1,bin(3),bin(3>>1)
```
